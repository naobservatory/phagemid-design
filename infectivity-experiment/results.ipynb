```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/colony_counts.csv")
```

```python
data = data[['sample','cfu_per_mL']]
```

```python
# Updating zero values to 1, for the sake of the log graph
data['cfu_per_mL'] = data['cfu_per_mL'].replace(0,1)
```

```python
data
```

```python
## Plot

custom_colors = ['#26bfed', '#FFFFFF', '#20b224', '#4dcc23']

plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(6, 4))
plt.rcParams['text.usetex'] = True
plt.rcParams['svg.fonttype'] = 'none'

ax = sns.barplot(x='sample', y='cfu_per_mL', data=data, ci='sd', capsize=.1, 
            errwidth=1.0, errcolor='black', palette = custom_colors,
           order = ['F- HP','F- PIC','F+ HP','F+ PIC'])
plt.yscale('log')  
plt.ylim([1,1E12])

ax = plt.gca()
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
ax.spines['right'].set_color('white')
ax.spines['top'].set_color('white')

plt.xlabel('\n Group')
plt.ylabel('CFU/mL')

plt.savefig('infectivity_resutls.pdf')
plt.show()
```
